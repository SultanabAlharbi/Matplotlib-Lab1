```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
ls=[2000,2005,2010,2015,2020]
df=pd.DataFrame({"Profit":[0,4000,9000,10000,150000]})

plt.plot(ls);
plt.plot(df["Profit"]);
```

## Q2: Create pie chart with equal explode values.

```python
Names=["Sultan","Ali","Abdullah","Fahad"]
Profit=[18000,23000,14000,18000]

plt.pie(Profit,labels=Names,autopct="%1.1f%%");
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
year=["2000","2005","2010","2015","2020"]
df=pd.DataFrame({"revenue":[0,4000,9000,10000,150000]})

plt.scatter(year,df["revenue"],alpha=0.4,s=50,marker="o");
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
year=["2000","2005","2010","2015","2020"]
df=pd.DataFrame({"revenue":[0,4000,9000,10000,150000]})

plt.scatter(year,df["revenue"],alpha=0.4,s=50,marker="o");
plt.title("revenue by year",loc="Left",color="red");
plt.xlabel("Year",fontsize=16);
plt.ylabel("revenue",fontsize=16);
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(["2000","2005","2010","2015","2020"],[1,6,9,4,2],color="red",label="Ali")
plt.plot(["2000","2005","2010","2015","2020"],[0,2,4,7,11],c="blue",label="Sultan")
plt.plot(["2000","2005","2010","2015","2020"],[10,20,30,40,50],"green",label="Abdullah")
plt.legend(loc="center left") ## Must set loc to "best"
plt.title("Employee Performance by year");
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)

#First
plt.subplot(2,3,1)
Names=["Sultan","Ali","Abdullah","Fahad"]
Profit=[18000,23000,14000,18000]

plt.pie(Profit,labels=Names,autopct="%1.1f%%");

#Second
plt.subplot(2,3,2)
year=["2000","2005","2010","2015","2020"]
df=pd.DataFrame({"revenue":[0,4000,9000,10000,150000]})
plt.scatter(year,df["revenue"],alpha=0.4,s=50,marker="o");
plt.title("revenue by year",loc="Left",color="red");
plt.xlabel("Year",fontsize=16);
plt.ylabel("revenue",fontsize=16);

##Third
plt.subplot(2,3,3)
plt.plot(["2000","2005","2010","2015","2020"],[1,6,9,4,2],color="red",label="Ali")
plt.plot(["2000","2005","2010","2015","2020"],[0,2,4,7,11],c="blue",label="Sultan")
plt.plot(["2000","2005","2010","2015","2020"],[10,20,30,40,50],"green",label="Abdullah")
plt.legend(loc="center left") ## Must set loc to "best"
plt.title("Employee Performance by year");

```
